```python
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

%matplotlib inline  
```

1. Standardize the d-dimensional dataset
2. Construct the covariance matrix
3. Decompose the convariance matrix into its eigenvectors and eigenvalues
4. Select k  eigenvectors that correspond to the k-largest eigenvalues, where k is the dimensionality of the new feature subspace


```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
```

```python
df_wine
```

```python
X = df_wine.iloc[:, 1:]
y = df_wine.iloc[:, 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state = 0)
```

Standardize the d-dimensional dataset

```python
sc = StandardScaler()
```

```python
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.fit_transform(X_test)
```

Construct the covariance matrix

```python
import numpy as np
cov_mat = np.cov(X_train_sc.T) # compute covariance matrix
```

Decompose the convariance matrix into its eigenvectors and eigenvalues

```python
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
```

```python
print eigen_vals
```

```python
total = sum(eigen_vals)
```

```python
var_exp = [(i / total) for i in sorted(eigen_vals, reverse=True)]
```

```python
cum_var_exp = np.cumsum(var_exp) # calculate the cumulative sum of explaned variances
```

```python
cum_var_exp
```

#### Plotting the variance explained ratios of the eigenvalues

Since the eigenvalues, define the magnitude of the eigenvectors, we sorted the eigenvalues by decreasing magnitude. Here, we're interested in the top K eigenvectors based on the values of ther corresponding eigenvalues



```python
plt.bar(range(1,14), var_exp, alpha= 0.5, label='Individual Explained Variance' )
plt.step(range(1,14), cum_var_exp, label='Cumulative Explained Variance')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.legend()
```
